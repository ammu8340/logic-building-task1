```python
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0
for m in marks:
    if m >= 50:
       print(f"{m} -> Pass")
       pass_count += 1
    else:
       print(f"{m} -> Fail")
       fail_count += 1
print("\nFinal Result:")
print("Total pass:", pass_count)
print("Total Fail:', fail_count)
```
